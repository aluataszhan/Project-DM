## 1)What is the average age and percentage of road accidents for all drivers based on genders? 

```python
average_age = ds[['Пол','Возраст, полных лет']].groupby(['Пол'], as_index=False).mean()
```

### Average age

```python
average_age
```

### Percentage

```python
without_acc = ds.groupby(['Пол'])['Регион ДТП'].apply(lambda x: x.isnull().sum())
with_accidents = ds.groupby('Пол')['Регион ДТП'].count()
percentage_accidents = (with_accidents/(with_accidents+without_acc))*100
```

```python
percentage_accidents
```

## 2)What is the difference between average driving experience and mean bonus-malus class for males and females?  

```python
avg_driving_exp = ds.groupby(['Пол'])['Стаж вождения'].mean()
mean_b_m = ds.groupby(['Пол'])['КБМ'].mean()
difference = avg_driving_exp - mean_b_m
difference
```

## 3)In which year the most significant number of vehicles were built? Please provide the list of top-10 years sorting based on the number of vehicles (descending). 

```python
ds.dtypes
#first we check datatype of column
```

```python
ds['Год выпуска'] = pd.to_datetime(ds['Год выпуска'], errors='coerce')
ds['Year'] = ds['Год выпуска'].dt.year
#we see that datatype of this column is "object" so we convert it. Then we separate the year from this date
```

```python
top10 = ds[['Year','Unique number']].groupby(['Year'], as_index=False).count().reset_index().rename(columns={'Unique number': 'Number of vehicles'})
top10.sort_values(by='Number of vehicles', ascending=False).head(10)
#count and finding top 10 by their number of vehicles
```

## 5)What is the most popular vehicle-make company in any 5 cities? (choose the cities you like most)  

```python
city = ['Алматы', 'Нур-Султан', 'Усть-Каменогорск', 'Петропавловск', 'Актобе']
top_5_city = ds[ds['Город'].isin([i for i in city])]
top_5_company = top_5_city[['Город', 'Модель','Unique number']].groupby(['Город', 'Модель']).count().reset_index().rename(columns={'Unique number': 'Vehicle'})
top_5_company = top_5_company.groupby('Город').max()
top_5_company
#i choose city that i like , and then move it to new dataset. Then I find most popular company(model) that makes vehicles
```

## 7)What is the most popular vehicle-make company for females in Almaty, Shymkent and Nur-Sultan (Astana) respectively?

## 9) How many people had more than 1 car in one year? Is there any driver who had more than 1 car accident in one year?

### Had car more than one 

```python
counting_cars = ds.groupby(['Unique number', 'Возраст, полных лет'])['Тип ТС '].count().reset_index()
more_than_1 = counting_cars[counting_cars['Тип ТС '] > 1].reset_index().rename(columns={'Тип ТС ': 'Number of cars'})
more_than_1
```

### had more than 1 car accident 

```python
counting_accidents = ds.groupby(['Unique number', 'Возраст, полных лет'])['Регион ДТП'].count().reset_index()
more_than_1_acc = counting_accidents[counting_accidents['Регион ДТП'] > 1].reset_index().rename(columns={'Регион ДТП': 'Number of accidents'})
more_than_1_acc
```

## 10) What is the distribution of road accidents based on the bonus-malus class?

```python
distribution = ds[['КБМ','Регион ДТП']].groupby(by=('КБМ'),as_index = False).count()
distribution = distribution.rename(columns={'Регион ДТП': 'Number of road accidents'})
distribution
#counting road accidents by their bonus malus
```

```python
distribution.plot( x='КБМ', y='Number of road accidents', kind='hist')
```
